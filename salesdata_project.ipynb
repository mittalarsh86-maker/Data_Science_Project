```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv("/content/Sales Dataset.csv")

print(df.head())
print(df.info())
```

```python
df['Date'] = pd.to_datetime(df['Date'])
monthly_sales = df.groupby(
    df['Date'].dt.strftime('%Y-%m')
)['Total Amount'].sum()

print(monthly_sales)
```

```python
plt.figure(figsize=(12,6))

monthly_sales.plot(
    kind='line',
    marker='o'
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales Amount")

plt.xticks(rotation=45)

plt.show()
```

```python
top_products = df.groupby(
    'Product Category'
)['Quantity'].sum().sort_values(
    ascending=False
)

print(top_products.head(10))
```

```python
regional_sales = df.groupby(
    'Product Category'
)['Quantity'].sum()

print(regional_sales)
```

```python
profit_region = df.groupby(
    'Region'
)['Profit'].sum()

print(profit_region)
```
