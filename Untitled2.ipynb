```python
import pandas as pd

df = pd.read_csv("/content/Student.csv")

df['Average'] = df[
    ['Maths','Physics','Chemistry']
].mean(axis=1)

top10 = df.sort_values(
    by='Average',
    ascending=False
).head(10)

print("Top 10 Performers")
print(top10)
```

```python
import pandas as pd

df = pd.read_csv("/content/Student.csv")

subjects = ['Maths','Physics','Chemistry']

for sub in subjects:

    highest = df[sub].max()

    print(sub)
    print("Highest Score =", highest)
```

```python
df['Average'] = df[
    ['Maths','Physics','Chemistry']
].mean(axis=1)

def grade(avg):

    if avg >= 80:
        return 'A'

    elif avg >= 60:
        return 'B'

    elif avg >= 40:
        return 'C'

    else:
        return 'D'

df['Grade'] = df['Average'].apply(grade)

print(df[['Average','Grade']])
```
